# Neural Representation Explorer

Explore how high-dimensional neural spike activity forms low-dimensional structure.

**Pipeline:** spike simulation → firing rate features → dimensionality reduction → clustering → visualization

```python
import sys
sys.path.insert(0, '..')

from src.simulate_spikes import simulate_spikes
from src.compute_features import compute_firing_rates
from src.dimensionality import compute_pca, compute_umap
from src.clustering import cluster_states

import matplotlib.pyplot as plt
import numpy as np
```

## 1. Simulate Neural Spikes

Generate Poisson spike trains for 60 neurons over 1000 time steps.

```python
spikes, state_labels = simulate_spikes(n_neurons=60, n_timesteps=1000)
print(f"Spike matrix shape: {spikes.shape}")
print(f"Example (first 5 neurons, first 10 timesteps):")
print(spikes[:5, :10])
```

## 2. Compute Firing Rate Features

Convert spike counts into sliding-window firing rates.

```python
features = compute_firing_rates(spikes, sigma=10)
print(f"Feature matrix shape: {features.shape}")
print(f"Each row is a {features.shape[1]}-dimensional neural population state.")
```

## 3. Dimensionality Reduction

Project neural states into 2D using PCA and UMAP.

```python
pca_embedding = compute_pca(features, n_components=2)
umap_embedding = compute_umap(features)

print(f"PCA embedding shape: {pca_embedding.shape}")
print(f"UMAP embedding shape: {umap_embedding.shape}")
```

## 4. Cluster Neural States

Group neural population states into clusters using K-Means.

```python
labels = cluster_states(features, k=4)
print(f"Cluster labels shape: {labels.shape}")
print(f"Unique clusters: {np.unique(labels)}")
```

## 5. Visualization

Visualize the low-dimensional neural manifold colored by cluster identity.

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

axes[0].scatter(pca_embedding[:, 0], pca_embedding[:, 1], c=labels, cmap='viridis', s=10, alpha=0.7)
axes[0].set_title('PCA — Neural Population States')
axes[0].set_xlabel('PC 1')
axes[0].set_ylabel('PC 2')

axes[1].scatter(umap_embedding[:, 0], umap_embedding[:, 1], c=labels, cmap='viridis', s=10, alpha=0.7)
axes[1].set_title('UMAP — Neural Population States')
axes[1].set_xlabel('UMAP 1')
axes[1].set_ylabel('UMAP 2')

plt.tight_layout()
plt.savefig('../neural_population_states.png', dpi=150, bbox_inches='tight')
plt.show()
```
